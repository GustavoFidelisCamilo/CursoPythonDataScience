```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = [5,10,15,20,25]
Y = [15,25,45,55,60]
plt.plot(X,Y)
plt.bar(X,Y)
plt.scatter(X,Y)
plt.show
```

```python
X = [5,10,15,20,25]
Y = [15,25,45,55,60]
fig,axs = plt.subplots(figsize=(12,4))
axs.set_xlabel('Eixo X')
axs.set_ylabel('Eixo Y')
axs.set_title('Funcao')
axs.plot(X,Y,"ro--", label ="Dados 1")
axs.plot(Y,X,"g>--",label ="Dados 2")
axs.legend()
```
